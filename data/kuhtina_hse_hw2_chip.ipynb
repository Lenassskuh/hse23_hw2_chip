Следует выбрать одну хромосому, потому что ресурсы google colab ограничены. В случае проблем, таких как Segmentation fault, error 139, следует выбрать хромосому меньшего размера.



Возьмем 17 хромосому для выравнивания

```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr17.fa.gz
!gzip -d chr17.fa.gz
```

```python
%%time
!bowtie2-build chr17.fa chromosome_index
```

```python
!mkdir bowtie2_res
```

```python
# 1 реплика
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF038JDF.fastq \
          -S bowtie2_res/res_JDF.sam
```

```python
# 2 реплика
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF723SKE.fastq \
          -S bowtie2_res/res_SKE.sam
```

```python
# контроль
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF051RXP.fastq \
          -S bowtie2_res/res_RXP.sam
```
